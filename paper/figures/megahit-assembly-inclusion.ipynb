```python
plass_d = {}
for row in df2.itertuples():
    plass_d['hu-' + row.genome] = float(row.plass_map)
```

```python
import sys, os
def latex_output(mincounts, fp=sys.stdout):
    for row in df.sort_values('similarity').itertuples():
        containment = float(row.similarity)
        num = os.path.basename(row.name)[:11]
        abbreviation = names_d[num]
        
        cont = '{:.1f}\\%'.format(containment * 100)
        plass_num = plass_d[num]
        plass_m = '{:.1f}\\%'.format(plass_num)

        print(abbreviation.replace('_', '\\_'), '&', cont, '&', plass_m, '\\\\', file=fp)

#latex_output(recA, open('recA-table.tex', 'wt'))
latex_output(df, open('../tables/megahit-inclusion.table.tex', 'wt'))
```
