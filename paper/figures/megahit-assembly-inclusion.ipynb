```python
import pandas
```

```python
names_df = pandas.read_csv('plass-genes/TR-hu-sb1-plass-gyrA.csv')

names_d = {}
for row in names_df.itertuples():
    names_d[row.name] = row.abbreviation
```

```python
df = pandas.read_csv('megahit-containment.csv', header=None, names=['containment', 'genome', 'reads', 'md5'])
```

```python
import sys
def latex_output(mincounts, fp=sys.stdout):
    for row in df.sort_values('containment').itertuples():
        containment = row.containment
        num = row.genome[:11]
        abbreviation = names_d[num]
        
        cont = '{:.1f}\\%'.format(containment * 100)

        print(abbreviation.replace('_', '\\_'), '&', cont, '\\\\', file=fp)

#latex_output(recA, open('recA-table.tex', 'wt'))
latex_output(df, open('../tables/megahit-inclusion.table.tex', 'wt'))
```
